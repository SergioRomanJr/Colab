```python
best_loss = 1000000000 # the number of values we want
best_m = 0
best_b = 0 

for m_guess in np.linspace(-100,100,100): # used for limiting the amount of information, for example: columns
    for b_guess in np.linspace(-100,100,100):
        
        y_guess = m_guess * x + b_guess
# making these guesses is going give best line of fit
        loss = np.sum((y_guess - y)**2)

        if loss < best_loss:
            
            best_loss = loss
            best_m = m_guess
            best_b = b_guess 
```

```python
y_guess = best_m * x + best_b
# here we are making a red best fit line
plt.plot(x,y_guess,'r');
plt.plot(D,C,'g.');
plt.show()
# plt. show and print is going to show the red line alongside its values
print(best_loss,best_m,best_b)
```
